```python
import re
```

```python
name_regex = re.compile(r"(?P<Name>Dorian\sGray)")
dorian = open('Dorian_Gray.txt', 'r')
lines = dorian.readlines()
i = 0
for line in lines:
    names = re.findall(name_regex, str(line))
    if names:
        i = i + 1
        print(f"Occurence: {i}\n\t{line}")
```

```python
grey = open('Dorian_Grey.txt', 'a')
name_regex = re.compile(r"(?P<Name>Dorian\sGray)")
dorian = open('Dorian_Gray.txt', 'r')
lines = dorian.readlines()
i = 0
for line in lines:
    names = re.findall(name_regex, str(line))
    if names:
        names = re.sub(r"Gray", "Grey", str(line))
        print(f"Occurence: {i}\n\t{names}")
        grey.writelines(names)
    else:
        grey.writelines(line)
grey.close()
```
